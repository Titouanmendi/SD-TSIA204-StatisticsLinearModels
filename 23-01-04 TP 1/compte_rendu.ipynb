```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

# Question 1

```python
df = pd.read_csv('invest.txt', sep = " ", header = 0)
print(df)
```

## Question 2

```python
log_gnp = np.log(df["gnp"])
log_invest = np.log(df["invest"])
plt.plot(log_gnp,log_invest)
plt.title("Investment on GNP")
```

## Question 3

```python
n = len(log_gnp)

# Compute the averages
gnp_av = sum(log_gnp)/n
invest_av = sum(log_invest)/n

# Compute the standard deviations squared of the gnp and the investments
gnp_var = 0
invest_var = 0
for i in range(n) :
    gnp_var = gnp_var + (log_gnp[i]-gnp_av)**2
    invest_var = invest_var + (log_invest[i]-invest_av)**2

# Compute the slope
slope = 0
for i in range(n) :
    slope = slope + (log_gnp[i]-gnp_av)*(log_invest[i]-invest_av)
slope = slope/gnp_var

# Compute the intercept
intercept = invest_av-slope*gnp_av

# Compute sigma's unbiased estimator squared

sigma_ue = 0
for i in range(n):
    sigma_ue = sigma_ue + (log_invest[i]-intercept-slope*log_gnp[i])**2
sigma_ue = sigma_ue/(n-2)

# Compute the slope's and the intercept's standard deviations

slope_var = sigma_ue/gnp_var
intercept_var = sigma_ue*(1/n + (gnp_av**2)/gnp_var)

# Display the results
print("The slope = ", slope)
print("The intercept =", intercept)
print("Intercept's standard deviation is ", intercept_var)
print("Slope's standard deviation is ", slope_var)
```
